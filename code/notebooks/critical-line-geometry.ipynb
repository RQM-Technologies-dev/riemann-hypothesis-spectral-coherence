# Critical Line Geometry

This notebook illustrates the geometry of the critical line Re(s) = 1/2 in the complex plane,
the reflection involution s ↦ 1−s, its fixed-point locus, and the placement of known
nontrivial zeros of ζ.

> **Note:** This notebook is for illustration and exploration only. Geometric pictures
> do not constitute proofs and are not used to establish any mathematical claim in the manuscript.


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

plt.rcParams.update({'figure.dpi': 120, 'font.size': 12})

# Known zero imaginary parts (Odlyzko 1987)
KNOWN_ZEROS_T = [
    14.134725142, 21.022039639, 25.010857580,
    30.424876126, 32.935061588, 37.586178159,
    40.918719012, 43.327073281, 48.005150881, 49.773832478,
]
```

## 1. The Critical Strip and Critical Line

The critical strip is the region 0 < Re(s) < 1 in the complex plane.
The critical line Re(s) = 1/2 is the unique axis of the reflection s ↦ 1−s.


```python
fig, ax = plt.subplots(figsize=(7, 10))
t_max = 55

# Shade critical strip
ax.axvspan(0, 1, alpha=0.08, color='blue', label='Critical strip 0 < Re(s) < 1')

# Critical line
ax.axvline(0.5, color='red', linewidth=2, label='Critical line Re(s) = 1/2')

# Strip boundaries
ax.axvline(0, color='gray', linewidth=1, linestyle='--')
ax.axvline(1, color='gray', linewidth=1, linestyle='--')

# Known zeros (on the critical line)
ax.scatter([0.5] * len(KNOWN_ZEROS_T), KNOWN_ZEROS_T,
           color='red', s=60, zorder=5, label='Known nontrivial zeros')

# Annotate first few zeros
for i, t0 in enumerate(KNOWN_ZEROS_T[:5]):
    ax.annotate(f'  ρ_{i+1} = 1/2 + {t0:.3f}i', (0.5, t0), fontsize=9, va='center')

ax.set_xlim(-0.2, 1.2)
ax.set_ylim(0, t_max)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s) = t')
ax.set_title('Critical strip and known nontrivial zeros of ζ')
ax.legend(loc='upper right', fontsize=9)
plt.tight_layout()
plt.savefig('../../media/figures/critical-strip-zeros.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. The Reflection Involution s ↦ 1−s

The map s ↦ 1−s reflects every point in the complex plane about the critical line Re(s) = 1/2.
Points on the critical line are fixed; off-line points are paired with their reflections.


```python
fig, ax = plt.subplots(figsize=(8, 6))

# Critical line
ax.axvline(0.5, color='red', linewidth=2, linestyle='-', label='Critical line Re(s) = 1/2', zorder=3)

# Illustrate reflection: a point s₀ and its image 1 − s₀
example_points = [
    (0.3, 20.0),
    (0.2, 35.0),
    (0.4, 10.0),
]

for (sigma0, t0) in example_points:
    s0 = complex(sigma0, t0)
    s1 = 1 - s0   # reflection: sigma -> 1 - sigma, t -> t

    ax.plot([sigma0, s1.real], [t0, s1.imag], 'b--', linewidth=1, alpha=0.6)
    ax.scatter([sigma0], [t0], color='blue', s=60, zorder=5)
    ax.scatter([s1.real], [s1.imag], color='green', s=60, zorder=5, marker='s')
    ax.annotate(f's₀ = {sigma0}+{t0}i', (sigma0, t0), textcoords='offset points',
                xytext=(-60, 5), fontsize=8, color='blue')
    ax.annotate(f'1−s₀ = {s1.real}+{s1.imag}i', (s1.real, s1.imag),
                textcoords='offset points', xytext=(5, 5), fontsize=8, color='green')

# On-line point (fixed)
ax.scatter([0.5], [25.0], color='red', s=80, zorder=6, marker='*')
ax.annotate('s = 1/2 + 25i\n(fixed by reflection)', (0.5, 25.0),
            textcoords='offset points', xytext=(5, 5), fontsize=8, color='red')

blue_patch = mpatches.Patch(color='blue', label='s₀ (off critical line)')
green_patch = mpatches.Patch(color='green', label='1−s₀ (reflection of s₀)')
red_patch = mpatches.Patch(color='red', label='Fixed points (on critical line)')
ax.legend(handles=[blue_patch, green_patch, red_patch], fontsize=9)

ax.set_xlim(0, 1)
ax.set_ylim(5, 45)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('The reflection s ↦ 1−s and its fixed-point locus')
plt.tight_layout()
plt.savefig('../../media/figures/reflection-involution.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Symmetry of Nontrivial Zeros

By the functional equation, zeros come in pairs {s, 1−s}. For zeros on the critical line
(Re(s) = 1/2), both elements of the pair coincide.

If a zero existed off the critical line at s₀ = σ₀ + it₀ with σ₀ ≠ 1/2, then 1−s₀,
s̄₀, and 1−s̄₀ would also be zeros — forming a quadruple. This notebook illustrates
this symmetry structure.


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Left: zeros on the critical line (the observed case)
ax = axes[0]
ax.axvline(0.5, color='red', linewidth=2, label='Re(s) = 1/2')
ax.axhline(0, color='gray', linewidth=0.5)
t0 = 21.022
# On-line zero and its pair (same point, reflected pair coincides)
ax.scatter([0.5, 0.5], [t0, -t0], color='red', s=100, zorder=5)
ax.annotate(f'ρ = 1/2 + {t0}i', (0.5, t0), xytext=(0.55, t0), fontsize=10, color='red')
ax.annotate(f'ρ̄ = 1/2 − {t0}i', (0.5, -t0), xytext=(0.55, -t0), fontsize=10, color='red')
ax.set_xlim(0, 1)
ax.set_ylim(-30, 30)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title('Zeros on the critical line:\nthe pair {ρ, 1−ρ} coincides')
ax.legend()

# Right: hypothetical off-line zero (not observed)
ax = axes[1]
ax.axvline(0.5, color='red', linewidth=2, linestyle='--', label='Re(s) = 1/2', alpha=0.5)
ax.axhline(0, color='gray', linewidth=0.5)
sigma0, t0 = 0.3, 21.022
quads = [
    (sigma0, t0, 'sₒ'),
    (1 - sigma0, t0, '1−sₒ'),
    (sigma0, -t0, 's̄ₒ'),
    (1 - sigma0, -t0, '1−s̄ₒ'),
]
for (s_re, s_im, label) in quads:
    ax.scatter([s_re], [s_im], color='purple', s=100, zorder=5)
    ax.annotate(label, (s_re, s_im), xytext=(s_re + 0.02, s_im + 0.5), fontsize=10, color='purple')
ax.set_xlim(0, 1)
ax.set_ylim(-30, 30)
ax.set_xlabel('Re(s)')
ax.set_ylabel('Im(s)')
ax.set_title(f'Hypothetical off-line zero at Re(s) = {sigma0}:\nquadruple symmetry {{{"sₒ, 1−sₒ, s̄ₒ, 1−s̄ₒ"}}}')
ax.legend()

plt.suptitle('Symmetry of nontrivial zeros: observed (left) vs hypothetical off-line (right)',
             fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig('../../media/figures/zero-symmetry.png', dpi=150, bbox_inches='tight')
plt.show()
```
