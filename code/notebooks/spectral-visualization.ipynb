# Spectral Visualization

This notebook visualizes the Riemann zeta function along and near the critical line,
illustrating the reflection symmetry ξ(s) = ξ(1−s) and the placement of known nontrivial zeros.

> **Note:** This notebook is for illustration and exploration only. Numerical computations
> do not constitute proofs and are not used to establish any mathematical claim in the manuscript.


```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import matplotlib.pyplot as plt
import mpmath
from placeholder import set_precision, zeta, xi, hardy_z, KNOWN_ZEROS_IMAGINARY_PARTS

set_precision(30)
mpmath.mp.dps = 30

plt.rcParams.update({'figure.dpi': 120, 'font.size': 12})
```

## 1. |ζ(1/2 + it)| Along the Critical Line

The Hardy Z-function Z(t) = e^{iθ(t)} ζ(1/2 + it) is real-valued and vanishes
exactly at the nontrivial zeros on the critical line.


```python
t_vals = np.linspace(0.1, 50, 2000)
Z_vals = [float(mpmath.siegelz(t)) for t in t_vals]

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(t_vals, Z_vals, color='steelblue', linewidth=0.8, label='Z(t)')
ax.axhline(0, color='black', linewidth=0.5)

# Mark known zeros
for t0 in KNOWN_ZEROS_IMAGINARY_PARTS:
    if t0 <= 50:
        ax.axvline(t0, color='red', linewidth=0.7, alpha=0.5)

ax.set_xlabel('t  (imaginary part of s = 1/2 + it)')
ax.set_ylabel('Z(t)')
ax.set_title('Hardy Z-function along the critical line Re(s) = 1/2')
ax.legend()
plt.tight_layout()
plt.savefig('../../media/figures/hardy-z-function.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. |ξ(s)| and the Reflection Symmetry

The completed zeta function ξ(s) satisfies ξ(s) = ξ(1−s). We verify this
numerically and visualize |ξ(s)| in the critical strip.


```python
# Verify functional equation numerically at several points
print("Functional equation check |ξ(s) - ξ(1-s)|:")
test_points = [
    complex(0.5, 14.134725),   # first zero
    complex(0.3, 10.0),        # off-line
    complex(0.7, 10.0),        # symmetric off-line
    complex(0.5, 30.0),
]
for s in test_points:
    diff = abs(xi(s) - xi(1 - s))
    print(f"  s = {s:12}  |ξ(s) - ξ(1-s)| = {float(diff):.2e}")
```

```python
# Plot |ξ(1/2 + it)| along the critical line
t_vals = np.linspace(0.1, 50, 1000)
xi_vals = [abs(float(mpmath.nstr(xi(complex(0.5, t)), 10))) for t in t_vals]

fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(t_vals, xi_vals, color='darkorange', linewidth=0.9, label='|ξ(1/2 + it)|')
ax.axhline(0, color='black', linewidth=0.5)

for t0 in KNOWN_ZEROS_IMAGINARY_PARTS:
    if t0 <= 50:
        ax.axvline(t0, color='red', linewidth=0.7, alpha=0.5)

ax.set_xlabel('t')
ax.set_ylabel('|ξ(1/2 + it)|')
ax.set_title('|ξ(s)| along the critical line Re(s) = 1/2')
ax.legend()
plt.tight_layout()
plt.savefig('../../media/figures/xi-critical-line.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. |ζ(σ + it)| in the Critical Strip

We compute |ζ(s)| on a grid over the critical strip 0 < Re(s) < 1, t ∈ [10, 40].
Zeros appear as dark regions (small |ζ|).


```python
sigma_vals = np.linspace(0.1, 0.9, 50)
t_vals_grid = np.linspace(10, 40, 200)

Z_grid = np.zeros((len(sigma_vals), len(t_vals_grid)))
for i, sigma in enumerate(sigma_vals):
    for j, t in enumerate(t_vals_grid):
        try:
            val = float(abs(mpmath.zeta(complex(sigma, t))))
            Z_grid[i, j] = val
        except Exception:
            Z_grid[i, j] = float('nan')

fig, ax = plt.subplots(figsize=(12, 5))
vmax = np.nanpercentile(Z_grid, 95)
im = ax.pcolormesh(t_vals_grid, sigma_vals, np.clip(Z_grid, 0, vmax),
                   cmap='viridis_r', shading='auto')
ax.axhline(0.5, color='red', linewidth=1.5, linestyle='--', label='Re(s) = 1/2')
plt.colorbar(im, ax=ax, label='|ζ(s)|')
ax.set_xlabel('t = Im(s)')
ax.set_ylabel('σ = Re(s)')
ax.set_title('|ζ(σ + it)| in the critical strip — zeros appear as dark regions')
ax.legend()
plt.tight_layout()
plt.savefig('../../media/figures/zeta-critical-strip.png', dpi=150, bbox_inches='tight')
plt.show()
```
